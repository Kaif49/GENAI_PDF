```python
import pandas as pd

def read_excel_data(file_path):
    # Load the Excel file
    df = pd.read_excel(file_path)
    # Convert the dataframe to a string format
    return df.to_csv(index=False)  # or use df.to_string() for a string representation

def query_data_with_gpt(data, question):
    response = openai.ChatCompletion.create(
        engine="aihackthon_turbo",  # Replace with the actual deployment name
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": f"Given the following PDF content, answer the question: {question}\n\n{data}"}
        ],
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.5,
    )
    return response['choices'][0]['message']['content'].strip()


# Load your Excel data
excel_path = "Devopsvul (1).xlsx"  # Replace with your Excel file path
data = read_excel_data(excel_path)
i=1 ;
# Define your query
question = "what is the {}th line in message coloumn".format(i)  # Replace with your actual query

# Get response from GPT-3.5
response = query_data_with_gpt(data, question)

# Print the response
print(response)

model = "gpt-35-turbo"
##prompt = ""

response = openai.ChatCompletion.create(
    engine="aihackthon_turbo",  # Note: use the engine name specific to your Azure deployment
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "solution for {}".format(response)}
    ],
    max_tokens=1024,
    n=1,
    stop=None,
    temperature=0.5,
)

print(response['choices'][0]['message']['content'])
```
