```python
import pandas as pd
import openai

# Function to read Excel data
def read_excel_data(file_path):
    # Load the Excel file
    df = pd.read_excel(file_path)
    # Convert the dataframe to a string format
    return df.to_csv(index=False)  # or use df.to_string() for a string representation

# Function to query the Excel data using GPT
def query_data_with_gpt(data, question):
    response = openai.ChatCompletion.create(
        engine="aihackthon_turbo",  # Replace with the actual deployment name
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": f"Given the following Excel content, answer the question: {question}\n\n{data}"}
        ],
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.5,
    )
    return response['choices'][0]['message']['content'].strip()

# Function to capture and save the output
def capture_gemini_output(output, filename="gemini_output.txt"):
    with open(filename, "a") as f:  # Open in append mode
        f.write(output + "\n")  # Add a newline after each output

# Load your Excel data
excel_path = "Devopsvul.xlsx"  # Replace with your Excel file path
data = read_excel_data(excel_path)

# Loop through i values from 2 to 10
for i in range(2, 5):
    # Define your query for each iteration
    question = "what is the {}th line in message column".format(i)

    # Get the initial response from GPT-3.5
    initial_response = query_data_with_gpt(data, question)
    print(f"Initial response for i={i}: {initial_response}")

    # Use the response to ask a follow-up question
    follow_up_response = openai.ChatCompletion.create(
        engine="aihackthon_turbo",  # Use the correct engine name
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": f"solution for {initial_response}"}
        ],
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.5,
    )['choices'][0]['message']['content']

    # Print and save the follow-up response
    print(f"Follow-up response for i={i}: {follow_up_response}")
    capture_gemini_output(follow_up_response)
```
