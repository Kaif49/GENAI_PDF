```python
import fitz
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# Function to query extracted text using GPT-3.5-turbo
def query_pdf_with_gpt(pdf_text, question):
    response = openai.ChatCompletion.create(
        engine="aihackthon_turbo",  # Replace with the actual deployment name
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": f"Given the following PDF content, answer the question: {question}\n\n{pdf_text}"}
        ],
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.5,
    )
    return response['choices'][0]['message']['content']

# Load your PDF
pdf_path = "TESTSQRRESULT.pdf"  # Replace with the actual PDF file path
pdf_text = extract_text_from_pdf(pdf_path)

# Query the PDF content
question = "how many status are there with OPEN"  # Replace with your query
response = query_pdf_with_gpt(pdf_text, question)

# Print the response from GPT-3.5-turbo
print(response)
```
